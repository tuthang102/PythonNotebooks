To assess our model's performance against our base line model 2.5*x, we can use bootstrap or permutation test to see if the two means of base line model and our model is statistically significant.

```python
observed_mean_diff = np.mean(2.5*x_train) - np.mean(2.7*x_train)
print('Observed difference in mean between two models:', abs(observed_mean_diff))
```

### Permutation test

```python
Nperm = 1000
mean_perm = []
k = 0 
for i in range(Nperm):
    pool = np.hstack([2.5*x_train, 2.7*x_train]) # combine the two groups
    np.random.shuffle(pool) # shuffle the combined data
    mean_diff = np.mean(pool[:10000]) - np.mean(pool[-10000:])
    mean_perm.append(mean_diff)
    if abs(mean_diff) >= abs(observed_mean_diff): # keeping track of how many times the mean difference is greater than the observed mean diff
        k += 1
```

```python
figure()
hist(mean_perm)
xlabel('Difference in mean')
ylabel('Count')
title('Permutation of mean')
show()
```

Since our p value is 0, we can conclude that the observed difference in the mean between two linear models is not due to chance. We reject the null hypothesis that the two means are the same.
